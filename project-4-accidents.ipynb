```python
X = df_clean02.drop(["Severity","State"], axis=1) #remove the column you don't need
y = df_clean02["Severity"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler #MinMaxScaler compress outliers
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
le = LabelEncoder()
le.fit(X_train['Sunrise_Sunset'].astype(str))
X_train['Sunrise_Sunset'] = le.transform(X_train['Sunrise_Sunset'].astype(str))
X_test['Sunrise_Sunset'] = le.transform(X_test['Sunrise_Sunset'].astype(str))
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
# label_encoder = LabelEncoder()
le.fit(y_train)
encoded_y_train = le.transform(y_train)
encoded_y_test = le.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train) 
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=8))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=60, activation='sigmoid'))
model.add(Dense(units=4, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=10,
    shuffle=True,
    verbose=2
)
```
